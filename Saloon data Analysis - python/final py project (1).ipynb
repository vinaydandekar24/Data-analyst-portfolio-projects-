## Problem Statement 01 : Understand different customer groups (based on spending, visits, and loyalty).

### Customer Segmentation Insights

```python
# Age distribution
plt.figure(figsize=(6,4))
sns.histplot(data['Age'], bins =25 ,kde=True)
plt.title("Age Distribution of Customers")
plt.show()
```

```python
# Gender distribution
plt.figure(figsize=(5,4))
sns.countplot(x='Gender', data=data )
plt.title("Customer Gender Distribution")
plt.show()
```

```python
# Visit Frequency Groups
plt.figure(figsize=(6,4))
sns.countplot(x='Visit_Frequency_Group', data=data, order=data['Visit_Frequency_Group'].value_counts().index)
plt.title("Visit Frequency Groups")
plt.show()
```

```python
# Customer Value distribution
plt.figure(figsize=(5,4))
sns.countplot(x='Customer_Value', data=data, order=['Low', 'Medium', 'High'])
plt.title("Customer Value Segments")
plt.show()
```

## Promotion & Loyalty Insights

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Loyalty_Program', hue='Promotional_Response', data=data)
plt.title("Promotional Response vs Loyalty Program")
plt.show()
```

```python
plt.figure(figsize = (4,4))
A = sns.histplot(data = df , x ='Total_Spend',kde = True  )
for i in A.containers:
    A.bar_label(i,fontweight = "bold", c = "k")
    plt.xticks(rotation = 0,fontweight = "bold")
    plt.yticks(rotation = 0,fontweight = "bold")
    plt.xlabel("Total_ Spend", fontweight = "bold", c = "k")
    plt.ylabel("Count", fontweight = "bold", c = "k")
    plt.show()
```

```python
sns.lmplot(data = df , x ='Total_Spend' ,y = "Average_Spend_Per_Visit", hue = "Gender" )
# this graph and histgram are shows the trends. now see the custombers are mostly spend the money in the group of 200 rs 
```

## Visit_Frequency_Group wise Total Spend

```python
plt.figure(figsize = [3,4])
explode = [0.2,0,0]
df.groupby("Visit_Frequency_Group")["Total_Spend"].sum().sort_values(ascending = False ).plot(kind = 'pie',explode = explode , startangle=90, autopct = "%.0f%%")
plt.title("Visit_Frequency_Group wise Total Spend ", fontweight  = 'bold')
plt.show()
```
